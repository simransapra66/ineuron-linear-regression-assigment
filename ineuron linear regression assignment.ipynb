### The task at hand is to build a linear regression model on the boston dataset to predict the price based on the dependent variable

```python
from sklearn.datasets import load_boston
boston=load_boston()
bos=pd.DataFrame(boston.data)
```

### Here for building our model we have to take into account the features which are important that means which have a high correlation for our model building that is 0.6 for the columns 2,5,6 and will build our model further on these features

```python
bos[[2]]
```

```python
from sklearn.model_selection import train_test_split
X = pd.DataFrame(np.c_[bos[[2]], bos[[5]],bos[[6]]], columns =['2','5','6'])
y=bos[[12]]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
print("The shape of the data after train test split is\n")
print(X_train.shape)
print(X_test.shape)
```

### Now we will construct our model of linear regression using scikit learn library

```python
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(X_train,y_train)
y_train_pred=lr.predict(X_train)
y_test_pred=lr.predict(X_test)
```

### The predictions of the train and test dataset are


```python
y_test
```

```python
y_test_pred
```

```python
print("The mean absolute error for this linear regression is\n")
m_s_e=mean_squared_error(y_test,y_test_pred)
print(m_s_e)
```

```python
from sklearn.metrics import r2_score
print("The R2 score for this is\n")
r2=r2_score(y_test,y_test_pred)
print(r2)
```
